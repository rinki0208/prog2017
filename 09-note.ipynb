1から6までの数をランダムに表示（1，6を含む）

```python
import random
rand_num = 1
while rand_num != 4:
    rand_num = random.randint(1,6)
    print(rand_num)
```

rand_numが4でない間繰り返す

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        break
print('.')
```

```python
import random
while True:
    rand_num = random.randint(1,6)
    print(rand_num)
    if rand_num == 4:
        continue
print('.')
```

```python
for x in range(1, 21):
    if x % 3 == 0:
        if x % 5 == 0:
            print('Fizz Buzz')   
        else:
            print(x)
    else:
        print(x)
```

```python
for x in range(1, 21):
    if x % 3 == 0:
        if x % 5 == 0:
            print('Fizz Buzz')   
        else:
            print('Fizz')
    elif x % 5 == 0:
        print('Buzz')
    else:
        print(x)
```

```python
for x in range(1,21):
    in x % 3 == 0 and x % 5 != 0:
        print('Fizz')
    in x % 3 != 0 and x % 5 == 0:
        print('Buzz')
    in x % 3 == 0 and x % 5 == 0:
        print('Fizz Buzz')
    in x % 3 != 0 and x % 5 != 0:
        print(x)
```
